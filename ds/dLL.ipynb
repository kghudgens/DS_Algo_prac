```python
class DoublyLinkedListNode(object):

    def __init__(self, value):

        self.value = value
        self.next_node = None
        self.prev_node = None
```

```python
a = DoublyLinkedListNode(1)
b = DoublyLinkedListNode(2)
c = DoublyLinkedListNode(3)
```

```python
a.next_node= b
b.prev_node = a
```

```python
b.next_node = c
c.prev_node = b
```
