```python
def func_constant(values):

    print(values[0])

    
```

```python
lst = [1,2,3]
func_constant(lst)
```

```python
def func_lin(lst):
    for val in lst:
        print(val)
```

```python
func_lin(list(range(1,11)))
```

```python
def func_quad(lst):

    for item_1 in lst:
        for item_2 in lst:
            print(item_1, item_2)
            
```

```python
func_quad(lst)
```
