```python
class Queue(object):
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.insert(0, item)

    def dequeue(self):
        return self.items.pop()

    def isEmpty(self):
        return self.items ==[]

    def size(self):
        return len(self.items)
```

```python
q = Queue()
```

```python
q.enqueue(3)
q.enqueue(33)
q.enqueue(32)
q.enqueue(356)
```

```python
q.size()
```

```python
q.dequeue()
```
